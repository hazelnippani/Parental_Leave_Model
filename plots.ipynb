```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"C:\Users\harin\PROJECTS\WECE\maternityModel\Prediction_Model\Parental_leave_2.csv", encoding='latin1')
data_sorted = data.sort_values(by='Paid Leave', ascending=False)
```

```python
industry_counts = data['Industry'].value_counts()
industry_top = industry_counts.head(15)
plt.figure(figsize=(7, 7))
plt.pie(industry_top, labels=industry_top.index, autopct='%1.1f%%')
plt.title('Top industries in DataSet')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

```python
#plot of maternity vs paternity leave
data = pd.read_csv(r"C:\Users\harin\PROJECTS\WECE\maternityModel\Prediction_Model\Parental_leave.csv", encoding='latin1')
data = data.interpolate()
dataa = np.array(data)
dataa = np.transpose(dataa)
X=dataa[2]
Y=dataa[4]
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions

plt.scatter(X, Y, s= 5, color = 'green')
plt.plot(X, Y_pred, color = 'blue')

plt.title('Maternity vs Paternity Paid Leave')
plt.xlabel('Maternity')
plt.ylabel('Paternity')
```
